```python
import matplotlib.pyplot as plt
import numpy as np

# Create some sample data
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure with two subplots (axes)
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(6, 8))

# Plot data on the first subplot (ax1)
ax1.plot(x, y1, label='sin(x)', color='blue')
ax1.set_title('Sine Wave')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()

# Plot data on the second subplot (ax2)
ax2.plot(x, y2, label='cos(x)', color='red')
ax2.set_title('Cosine Wave')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.legend()

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a Figure and multiple Axes
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(6, 8))

# Plot on the first Axes (top subplot)
ax1.plot(x, y1, color='blue', label='sin(x)')
ax1.set_title('Sine Function')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')
ax1.legend()

# Plot on the second Axes (bottom subplot)
ax2.plot(x, y2, color='red', label='cos(x)')
ax2.set_title('Cosine Function')
ax2.set_xlabel('x')
ax2.set_ylabel('cos(x)')
ax2.legend()

# Add some space between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line plot
plt.plot(x, y)

# Add labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')

# Show the plot
plt.show()
```

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a scatter plot
plt.scatter(x, y, marker='o', color='blue', label='Data Points')

# Add labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Scatter Plot')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 30, 50]

# Create a bar chart
plt.bar(categories, values, color='green')

# Add labels and a title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Chart')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a sample dataset
data = sns.load_dataset("tips")

# Create a histogram
sns.histplot(data=data, x="total_bill")

# Add labels and title
plt.xlabel("Total Bill Amount")
plt.ylabel("Frequency")
plt.title("Histogram of Total Bill Amount")

# Show the plot
plt.show()
```
