```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(-180,180,60)
```

```python
y=np.sin(x)
z=np.cos(x)
```

# Sine Wave

```python
plt.figure(figsize=(7,7))
plt.plot(x,y)
plt.xlabel('Angels')
plt.ylabel('Values')
plt.title('Sin Wave')
plt.show()
```

# Cosine Wave

```python
plt.figure(figsize=(8,8))
plt.plot(x,z)
plt.xlabel('angles')
plt.ylabel('Values')
plt.title('Cosine Wave')
plt.show()
```

```python
x = np.linspace(-10,10,20)
y=x**2
plt.figure(figsize=(7,7))
plt.plot(x,y)
plt.xlabel('values')
plt.title('parabola')
plt.show()
```

# Bar Plot

```python
import numpy as np
import matplotlib.pyplot as plt
 
  
# creating the dataset
data = {'C':40, 'C++':25, 'Java':50,
        'Python':55}
courses = list(data.keys())
students = list(data.values())
  
plt.figure(figsize = (10, 5))
plt.xlabel('Courses')
plt.ylabel('Students')
 
# creating the bar plot
plt.bar(courses, students, color ='maroon',
        width = 0.4)
```

```python
plt.xlabel('students')
plt.ylabel('Courses')
plt.barh(courses, students, color='green')
```

# Pie Plot

```python
plt.figure(figsize=(7,7))
plt.pie(students, labels=courses, autopct='%1.1f%%')
plt.show()
```

# Scatter Plot

```python
x = np.linspace(0,10,30)
y=np.sin(x)
z= np.cos(x)
plt.figure(figsize=(7,7))
plt.scatter(x,y)
plt.scatter(x,z)
plt.title('Scatter Plot')
plt.show()
```

# 3D Scatter Plot

```python
plt.figure(figsize=(7,7))
plt.axes(projection='3d')
z=20* np.random.random(100)
x=np.sin(z)
y = np.cos(z)
plt.scatter(x,y,z, c=z, cmap='Blues')
plt.show()
```
