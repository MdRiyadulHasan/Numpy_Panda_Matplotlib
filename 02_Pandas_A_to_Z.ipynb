# Working with pandas DataFrame

```python
import pandas as pd
```

```python
from sklearn.datasets import load_boston
```

```python
boston=load_boston()
```

```python
boston
```

```python
boston_df=pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
boston_df.head()
```

```python
boston_df.tail()
```

```python
boston_df.shape
```

# Information about the dataset

```python
boston_df.info()
```

```python
boston_df.describe()
```

# Finding Number of Missing Values

```python
boston_df.isnull().sum()
```

# Adding new columns in the dataset

```python
boston_df['Price']=boston.target
```

```python
boston_df.head()
```

# Removing Row and Column from the dataset

```python
boston_df.drop(index=3, axis=0)
```

```python
boston_df.drop(columns='ZN', axis=1)
```

```python
boston_df.iloc[2]
```

```python
boston_df.iloc[:,2]
```

# Correlation - positive and negative

```python
boston_df.corr()
```

# Save DataFrame to csv file

```python
boston_df.to_csv('datasets/boston.csv')
```

# Save DataFrame to Excel file

```python
boston_df.to_excel('datasets/boston.xlsx')
```
